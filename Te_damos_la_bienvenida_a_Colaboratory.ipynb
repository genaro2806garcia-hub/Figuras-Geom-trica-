```
import math


class FigurasGeometricas:
    """Clase base que representa una figura geométrica con encapsulamiento."""

    def __init__(self, alto, ancho):
        self.__alto = alto
        self.__ancho = ancho

    @property
    def alto(self):
        """Getter: retorna el valor del alto."""
        return self.__alto

    @alto.setter
    def alto(self, valor):
        """Setter: valida y asigna el alto."""
        if valor <= 0:
            raise ValueError("El alto debe ser un valor positivo.")
        self.__alto = valor

    @property
    def ancho(self):
        """Getter: retorna el valor del ancho."""
        return self.__ancho

    @ancho.setter
    def ancho(self, valor):
        """Setter: valida y asigna el ancho."""
        if valor <= 0:
            raise ValueError("El ancho debe ser un valor positivo.")
        self.__ancho = valor

    def area(self):
        return self.__alto * self.__ancho

    def perimetro(self):
        return 2 * (self.__alto + self.__ancho)

    def __str__(self):
        return (f"Figura Geométrica | Alto: {self.__alto} cm | "
                f"Ancho: {self.__ancho} cm")


class Rectangulo(FigurasGeometricas):

    def __init__(self, alto, ancho):
        super().__init__(alto, ancho)

    def area(self):
        return self.alto * self.ancho

    def perimetro(self):
        return 2 * (self.alto + self.ancho)

    def __str__(self):
        return (f"Rectángulo       | Alto: {self.alto} cm | "
                f"Ancho: {self.ancho} cm | "
                f"Área: {self.area():.2f} cm² | "
                f"Perímetro: {self.perimetro():.2f} cm")


class Triangulo(FigurasGeometricas):

    def __init__(self, alto, ancho):
        super().__init__(alto, ancho)

    def area(self):
        return (self.ancho * self.alto) / 2

    def perimetro(self):
        hipotenusa = math.sqrt(self.alto**2 + self.ancho**2)
        return self.alto + self.ancho + hipotenusa

    def __str__(self):
        return (f"Triángulo        | Alto: {self.alto} cm | "
                f"Base: {self.ancho} cm | "
                f"Área: {self.area():.2f} cm² | "
                f"Perímetro: {self.perimetro():.2f} cm")


class Elipse(FigurasGeometricas):

    def __init__(self, alto, ancho):
        super().__init__(alto, ancho)

    def area(self):
        return math.pi * self.alto * self.ancho

    def perimetro(self):
        a, b = self.alto, self.ancho
        h = ((a - b) ** 2) / ((a + b) ** 2)
        return math.pi * (a + b) * (1 + (3 * h) / (10 + math.sqrt(4 - 3 * h)))

    def __str__(self):
        return (f"Elipse           | Semi-eje a: {self.alto} cm | "
                f"Semi-eje b: {self.ancho} cm | "
                f"Área: {self.area():.2f} cm² | "
                f"Perímetro: {self.perimetro():.2f} cm")


if __name__ == "__main__":
    separador = "=" * 65

    print(separador)
    print("       Práctica de Clase (POO en Python)")
    print(separador)

    figura = FigurasGeometricas(3, 7)
    print("\n[1] Figura base creada:")
    print(f"    alto  (getter) → {figura.alto} cm")
    print(f"    ancho (getter) → {figura.ancho} cm")

    figura.alto = 11
    figura.ancho = 9
    print("\n[2] Valores actualizados con setters:")
    print(f"    alto  (setter) → {figura.alto} cm")
    print(f"    ancho (setter) → {figura.ancho} cm")
    print(f"    Área base      → {figura.area():.2f} cm²")
    print(f"    Perímetro base → {figura.perimetro():.2f} cm")

    print("\n[3] Prueba del setter con valor inválido:")
    try:
        figura.ancho = -3
    except ValueError as e:
        print(f"     Error capturado correctamente: {e}")

    print(f"\n{separador}")
    print("       HERENCIA Y SOBRESCRITURA DE MÉTODOS")
    print(separador)

    rect = Rectangulo(8, 14)
    tri  = Triangulo(6, 5)
    eli  = Elipse(10, 4)

    print(f"\n  {rect}")
    print(f"  {tri}")
    print(f"  {eli}")

    print(f"\n[4] Cambiar dimensiones del rectángulo con setter:")
    rect.alto  = 13
    rect.ancho = 7
    print(f"  {rect}")

    print(f"\n{separador}")
    print("  Programa ejecutado sin errores   ")
    print(separador)
```
